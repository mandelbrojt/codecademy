```python
# Replace each ";,;" with the "%" character
daily_sales_replaced = daily_sales.replace(";,;", "%")

# Split the replaced string by each ","
daily_transactions = daily_sales_replaced.split(",")

# Create an empty list to store each transaction
daily_transactions_split = []

# For each transaction split the string and append it to the empty list created previously
for element in daily_transactions:
    daily_transactions_split.append(element.split("%"))

# Create an empty list to store the stripped transactions
transactions_clean = []

# For each transaction data strip white spaces
for transaction in daily_transactions_split:
    for data in transaction:
        transactions_clean.append(data.strip())

# Create an empty list to store customers names
customers = []

# Create an empty list to store sales amounts
sales = []

# Create an empty list to store thread data
thread_sold = []

# Loop through cleaned transaction and extract each customer's name
for i in range(0, len(transactions_clean), 4):
    customers.append(transactions_clean[i])

# Loop through cleaned transactions and extract each sale
for i in range(1, len(transactions_clean), 4):
    sales.append(transactions_clean[i])

# Loop through cleaned transactions and extract each thread
for i in range(2, len(transactions_clean), 4):
    thread_sold.append(transactions_clean[i])

# Initialize the total sales variable
total_sales = 0

# Loop through each item in sales and remove the "$" sign
# Convert each item into a float and add it to total sales
for sale in sales:
    total_sales += float(sale.replace("$", ""))

# Create an empty list to store characters splitted by "&"
splitted_by_amper = []

# Create an empty list to store sliced colors
thread_sold_split = []

# Loop through each element of thread_sold
for thread in thread_sold:
    # Check if "&" is in each element of the loop
    if "&" not in thread:
        # Append the color name to thread_sold_split
        thread_sold_split.append(thread)
    else:
        # Access each character of the string
        for letter in thread:
            # Check if there is any "&" on each character
            if "&" in letter:
                # Split the character by each "&"
                splitted_by_amper = thread.split(letter)
                # Loop through each splitted character
                for i in splitted_by_amper:
                    # Add each color name to thread_sold_split
                    thread_sold_split.append(i)
            
# Create an empty list to store unique colors
unique_colors = []

# For each element on splitted colors
for color in thread_sold_split:
    # Check if color is already on unique_colors
    if color not in unique_colors:
        # If the color is not already stored append it to unique_colors
        unique_colors.append(color)

def color_count(color):
    count = 0
    for thread in thread_sold_split:
        if thread == color:
            count += 1
    return count

for i in range(len(unique_colors)):
    print("There were {threads} {color} threads sold today".format(threads=color_count(unique_colors[i]), color=unique_colors[i]))
```
