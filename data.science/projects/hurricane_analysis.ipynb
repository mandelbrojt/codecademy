```python
# 1
# Update Recorded Damages

""" Write a function that returns a new list of updated damages where 
the recorded data is converted to float values and the missing data 
is retained as "Damages not recorded".
"""

conversion = {"M": 1000000,
              "B": 1000000000}

def convert_damages(damages_list):
    # Create an empty lists to append values by conditionals
    damages_converted = []
    
    # Iterate through each element of the function parameter
    for element in damages_list:
        
        # Test if element is equal to a specific string
        if element == "Damages not recorded":
            # Add the value of element to the empty list
            damages_converted.append(element)
        
        # Tests if the character "M" is in element
        elif "M" in element:
            # Slice the character to select only number characters
            # Convert each slice into a float number
            damages_converted.append(float(element[:element.find("M")]) * conversion["M"])
        
        # Tests if the character "B" is in element
        elif "B" in element:
            # Slice the character to select only number characters
            # Convert each slice into a float number
            damages_converted.append(float(element[:element.find("B")]) * conversion["B"])
    
    return damages_converted

damages_cleaned = convert_damages(damages)
```

```python
# 2
# Create a Table

"""
Write a function that constructs a dictionary made out of the lists,
where the keys of the dictionary are the names of the hurricanes,
and the values are dictionaries themselves containing a key for 
each piece of data (Name, Month, Year,Max Sustained Wind, 
Areas Affected, Damage, Death) about the hurricane.
"""

def organize_by_name():
    # Empty dictionary to store values of the main dictionary
    values_dict = {}
    # Empty dictionary to store a dictionary as values
    main_dict = {}
    
    # Iterate by the length of the "names" list
    for i in range(len(names)):
        # Set each dictionary key as the value of "i"
        # Set each dictionary value as the "i" element of each specified list
        values_dict[i] = {"Name": names[i], "Month": months[i],
                          "Year": years[i], "Max Sustained Wind": max_sustained_winds[i],
                          "Areas Affected":areas_affected[i], "Damage": damages_cleaned[i],
                          "Deaths": deaths[i]}

    # Iterate through the items of "values dictionary"
    for key,value in values_dict.items():
        # Set each key as the "ith" element of the "names" list
        # Set the value of "values_dict" as the value of "main_dict"
        main_dict[names[key]] = value
    return main_dict

hurricanes_by_name = organize_by_name()
```

```python
# 3
# Organizing by Year

""" Write a function that converts the current dictionary of hurricanes to a 
new dictionary, where the keys are years and the values are lists containing 
a dictionary for each hurricane that occurred in that year.
"""

def organize_by_year():
    # Create an empty list to store "unique_years" values
    unique_years = []

    # Iterate through each value of "years" list
    for year in years:
        # If "year" is NOT IN "unique_years"
        if year not in unique_years:
            # Append the "year" to "unique_years"
            unique_years.append(year)

    # create a dictionary to store the names of the hurricanes per year
    # key = Unique year
    # value = Hurricane name
    year_hurricanes = {}

    # Iterate through each year in "unique_years"
    for year in unique_years:
        # Create an empty list to store hurricane names
        hurricanes_in_year = []
        # Iterate through each key and value of "hurricanes" dict
        for key, value in hurricanes_by_name.items():
            # If "year" is equal to "Year" key in "hurricanes"
            if year == value["Year"]:
                # Append hurricane name to "hurricanes_in_year"
                hurricanes_in_year.append(key)
        # key = year
        # value = hurricanes_in_year list
        year_hurricanes[year] = hurricanes_in_year

    # Create an empty dict to store hurricana data for each year
    # key = year
    # value = hurricane data
    hurricanes_by_year = {}

    # Iterate through each key and value of the year_hurricanes dict
    for year, hurricanes in year_hurricanes.items():
        # Create an empty list to store each hurricane data per year
        hurricane_data = []
        # Iterate through each value of hurricanes
        for hurricane in hurricanes:
            # Append the hurricane data provided by the hurricanes_by_name dict
            hurricane_data.append(hurricanes_by_name[hurricane])
        # key = year
        # value = hurricane_data list
        hurricanes_by_year[year] = hurricane_data

    return hurricanes_by_year

hurricanes_by_year = organize_by_year()
    
```

```python
# 4
# Counting Damaged Areas

""" Write a function that counts how often each area is listed as an affected area 
of a hurricane. Store and return the results in a dictionary where the keys are the 
affected areas and the values are counts of how many times the areas were affected.

Test your function on your hurricane dictionary.
"""
def affected_areas_counter():
    # Create empty list to store unique values of areas
    unique_areas = []

    # Iterate through each key and value in "hurricanes_by_name"
    for hurricane, data in hurricanes_by_name.items():
        # Iterate through each single area value
        for area in data["Areas Affected"]:
            # If the value of area IS NOT in "unique_areas"
            if area not in unique_areas:
                # Append the value of area to "unique_areas"
                unique_areas.append(area)

    # create dictionary of areas to store the number of hurricanes involved in
    affected_dict = {}

    # Iterate through each area value in "unique_areas"
    for area in unique_areas:
        # Initialize a counter
        count = 0
        # Iterate through each value of "hurricanes_by_name"
        for value in hurricanes_by_name.values():
            # Iterate through each single value of "Areas Affected"
            for unique_area in value["Areas Affected"]:
                # If the unique area and the value of Areas Affected are equal
                if area == unique_area:
                    # Increase the counter by 1
                    count += 1
        # Assign the area as key
        # Assign the counter as value
        affected_dict[area] = count
    return affected_dict

areas_affected = affected_areas_counter()
```

```python
# 5 
# Calculating Maximum Hurricane Count

""" Write a function that finds the area affected by the 
most hurricanes, and how often it was hit.
"""
def most_affected_area():
    occurrences = []
    for count in areas_affected.values():
        occurrences.append(count)
    max_occurrence = max(occurrences)
    
    max_area = ""
    for area, count in areas_affected.items():
        if max_occurrence == count:
            max_area += area

    print("The most affected area is {area} with {count} occurences".format(area=max_area, count=max_occurrence))

# find most frequently affected area and the number of hurricanes involved in
most_affected_area()
```

```python
# 6
# Calculating the Deadliest Hurricane

""" Write a function that finds the hurricane that caused the 
greatest number of deaths, and how many deaths it caused.
"""
# Function that finds the dedliest hurricane
def most_deaths():
    deaths = []
    for value in hurricanes_by_name.values():
        deaths.append(value["Deaths"])
    max_death = max(deaths)
    
    deadliest_area = ""
    for key, value in hurricanes_by_name.items():
        if max_death == value["Deaths"]:
            deadliest_area += key

    print("The most dedliest area is {area} with {deaths} number of deaths".format(area=deadliest_area, deaths=max_death))

# find highest mortality hurricane and the number of deaths
most_deaths()
```

```python
# 7
# Rating Hurricanes by Mortality
def hurricanes_by_mortality():
    
    mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000,
                   5: 100000}
    
    moratality_dict = {}

    for i in range(5):
        names_in_range = []
        for name, death in zip(names, deaths):
            if death > mortality_scale[i] and death <= mortality_scale[i+1]:
                names_in_range.append(name)
                moratality_dict[i] = names_in_range
    
    return moratality_dict

hurricanes_by_mortality = hurricanes_by_mortality()
```

```python
# 8 Calculating Hurricane Maximum Damage

"""Write a function that finds the hurricane that 
caused the greatest damage, and how costly it was.
"""
def greatest_damage():
    float_costs = []
    for cost in damages_cleaned:
        if type(cost) == type(1.0):
            float_costs.append(cost)

    max_cost = max(float_costs)

    hurricane = ""

    for key, value in hurricanes_by_name.items():
        if value["Damage"] == max_cost:
            hurricane += key
    
    print(f"The highest damage by total cost was {hurricane} with a value of {max_cost}")

greatest_damage()
```

```python
# 9
# Rating Hurricanes by Damage

"""Write a function that rates hurricanes on a damage scale according 
to the following ratings, where the key is the rating and the value 
is the upper bound of damage for that rating.
"""
def hurricanes_by_severity():
    
    damage_scale = {0: 0,
                    1: 100000000.0,
                    2: 1000000000.0,
                    3: 10000000000.0,
                    4: 50000000000.0}
    
    hurricanes_filtered = []
    costs_filtered = []

    for hurricane, cost in zip(names, damages_cleaned):
        if type(cost) == type(1.0):
            hurricanes_filtered.append(hurricane)
            costs_filtered.append(cost)
    
    severity_dict = {}

    for i in range(4):
        names_in_range = []
        for hurricane, cost in zip(hurricanes_filtered, costs_filtered):
            if cost > damage_scale[i] and cost <= damage_scale[i+1]:
                names_in_range.append(hurricane)
                severity_dict[i] = names_in_range
    
    return severity_dict

hurricanes_by_severity = hurricanes_by_severity()
```
